### Penggabungan dataset

```python
# Gunakan dataset timbunan sampah menjadi dataset dasar
df = df_timbunanSampah

# Melihat data
df.head()
```

```python
# Bentuk list dari nama kolom data sumber sampah
sumberSampah_feature = ['Rumah Tangga(ton)',
       'Perkantoran(ton)', 'Pasar(ton)', 'Perniagaan(ton)',
       'Fasilitas Publik(ton)', 'Kawasan(ton)', 'Lain(ton)']

# Bentuk list dari nama kolom data jenis sampah
jenisSampah_feature = ['Sisa Makanan (%)',
       'Kayu-Ranting (%)', 'Kertas-Karton (%)', 'Plastik(%)', 'Logam(%)',
       'Kain(%)', 'Karet- Kulit (%)', 'Kaca(%)', 'Lainnya(%)']
```

```python
# Lakukan pemindahan kolom dan value data sumber sampah dan jenis sampah ke dalam dataframe utama

## Data sumber sampah

### Loop kolom feature sumber sampah
for col in sumberSampah_feature:
    ### Tambahkan setiap feature dan valuenya ke dalam df
    df[col] = df_sumberSampah[col]

## Data jenis sampah

### Loop kolom feature jenis sampah
for col in jenisSampah_feature:
    ### Tambahkan setiap feature dan valuenya ke dalam df
    df[col] = df_jenisSampah[col]
```

#### Imputasi data Jenis sampah

Imputasi data jenis sampah dilakukan dengan mempertimbangkan feature lainnya dikarenakan dalam feature tersebut bentuknya berupa %. Jadi apabila terdapat nan seperti pada row index 1 di bawah, dapat dilakukan penjumlahan feature yang tidak null value lalu dikurangkan 100 sehingga didapatkan nilai missing value yang hilang pada kolom Lainnya(%).

Index | Sisa Makanan (%) | Kayu-Ranting (%) | Kertas-Karton (%) | Plastik (%) | Logam (%) | Kain (%) | Karet-Kulit (%) | Kaca (%) | Lainnya (%) |
|------------------|------------------|------------------|-------------------|-------------|-----------|----------|-----------------|----------|--------------|
| 1                | 36.00            | 6.00              | 6.00        | 30.00     | 9.00     | 5.00            | 3.00     | 5.00         | Nan            

Secara lengkap rumus imputasi di atas

$$\text{Nan Value} = 100 - \sum_{i=1}^{n} \text{Non null feature}_i$$

Apabila dalam satu row terdapat >1 nan feature seperti pada data index 9 di bawah, maka nilai nan yang didapat menggunakan rumus diatas akan dibagi dengan jumlah row yang memiliki missing value.

Index | Sisa Makanan (%) | Kayu-Ranting (%) | Kertas-Karton (%) | Plastik (%) | Logam (%) | Kain (%) | Karet-Kulit (%) | Kaca (%) | Lainnya (%) |
|------------------|------------------|------------------|-------------------|-------------|-----------|----------|-----------------|----------|--------------|
| 9                | Nan            | Nan            | 24.60              | 5.54        | 12.30     | 6.16     | Nan            | 12.91     | 20.72         | 

Rumus imputasi apabila nan feature > 1: 

$$\text{Nan Value} = 100 - \sum_{i=1}^{n} \text{Non null feature}_i$$

$$\text{Nan Value per feature} = \frac{Nan Value} {n total}$$

```python
df[jenisSampah_feature]
```

```python
# Loop seluruh index
for i in range(len(df.index)):
    # Deklarasi seluruh feature yang kosong ke dalam list
    misscol = []
    # Loop seluruh feature jenis sampah
    for col in jenisSampah_feature:
        # Deteksi feature yang nan menggunakan if
        if pd.isnull(df.loc[i, col]) == True:
            # Apabila ada, tambahkan ke list misscol
            misscol.append(col)
    # Dilakukan complement untuk mengetahui kolom yang tidak kosong
    nonmisscol = list(set(jenisSampah_feature) - set(misscol))
    # Apabila feature yang kosong hanya 1
    if len(misscol)==1:
        # Dapatkan value feature yang tidak kosong ke dalam list
        total = df.loc[i, nonmisscol].tolist()
        # Total semua value dalam list
        total = sum(total)
        # Kurangkan hasil dengan 100
        total = 100-total
        # Imputasi nilai pada feature yang kosong
        df.loc[i, misscol] = total
    # Apabila feature yang kosong > 1
    elif len(misscol)>1:
        # Dapatkan value feature yang tidak kosong ke dalam list
        total = df.loc[i,nonmisscol].tolist()
        # Total semua value dalam list
        total = sum(total)
        # Kurangkan hasil dengan 100 lalu bagi dengan jumlah n feature yang kosong
        total = (100-total)/len(misscol)
        # Bulatkan 2 angka di belakang koma
        total = round(total,2)
        # Loop kolom yang merupakan missing coloumns
        for col in misscol:
            # Imputasi nilai pada feature yang kosong
            df.loc[i, misscol] = total
```
